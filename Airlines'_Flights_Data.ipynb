The boxplot displays how ticket prices vary with flight duration, which has been grouped into bins. Here's what we can observe:

**Median Price**: As the flight duration increases, the median ticket price generally tends to increase. We can see a clear upward trend in the horizontal line inside each box.

**Price Range and Variability**: Shorter duration flights (e.g., up to 5 hours) show a smaller spread in prices, meaning less variability. As duration increases, the boxes (representing the interquartile range, or the middle 50% of data) become taller, indicating a wider range of prices and greater variability for longer flights.

**Outliers**: There are several outliers, particularly at higher price points, across various duration bins. These represent unusually expensive flights for their respective duration categories.

In summary, the boxplot suggests a positive correlation: longer flight durations are generally associated with higher and more variable ticket prices.

*Question 7*: Analysis of Price distribution

```python
sns.histplot(df['price'],kde=True)
plt.xlabel("Price")
plt.ylabel("Frequency")
```

```python
print(f"Skewness of Price: {df['price'].skew():.2f}")
```

The histogram of flight prices, along with the KDE (Kernel Density Estimate) curve, provides insights into the distribution:

*   **Skewness**: The distribution is **right-skewed (positively skewed)**. This means that the tail of the distribution is longer on the right side, indicating that there are many flights with lower prices and fewer flights with very high prices. The calculated skewness value (likely positive) confirms this visual observation.

*   **Most Common Fare Range**: The highest peaks in the histogram are towards the lower price ranges. The most common fare ranges appear to be in the **0 - 10,000** and **40,000 - 50,000** (approximately) brackets, with a noticeable drop in frequency for prices in between.

*   **Outliers**: The long tail extending to the right clearly shows the presence of **outliers**, which are flights with significantly higher prices than the majority. These higher-priced flights contribute to the positive skewness. While the histogram shows their presence, a boxplot would be more effective for a detailed outlier analysis.

comparing expensive vs cheaper tickets

```python
median_price=df['price'].median()
df['Price_Category']=df['price'].apply(lambda x: 'Expensive' if x>median_price else 'Cheap')
```

```python
pd.crosstab(df['airline'],df['Price_Category'])
```

```python
pd.crosstab(df['Price_Category'],df['stops'])
```

```python
pd.crosstab(df['Price_Category'],df['airline'])
```

```python
cross=pd.crosstab(df['airline'],df['Price_Category'],normalize='index')*100
cross.round(2)
```

```python
cross.plot(kind='bar',stacked=True,figsize=(8,5))
plt.title('Proportion of Cheap and Expensive Tickets by Airline')
plt.xlabel("Airline")
plt.ylabel("Percentage")
plt.legend(title='Price Category')
```

Airlines such as Vistara and AirIndia have a significantly higher proportion of expensive tickets compared to other airlines. In contrast AirAsia, GOFirst, IndiGo and SpiceJet primary offer tickets in the lower price category. This suggests that the airline itself is an important factor associated with ticket pricing, possibly reflecting differences in service quality, route network or market positioning

```python
numeric=df.select_dtypes(include='number')
sns.heatmap(numeric.corr(),annot=True,cmap='coolwarm')
```

This heatmap displays the correlation matrix for the numerical columns: `days_left`, `price`, `actual_duration`, and `deparure_hour`.

*   **Price and Days_Left**: There appears to be a **a very negative correlation (around -0.092)** between `price` and `days_left`. This suggests that as the number of days left until departure increases, the price tends to decrease slightly. This is a common pattern where booking further in advance can result in lower prices.

*   **Price and Actual_Duration**: There is a **positive correlation (around 0.20)** between `price` and `actual_duration`. This indicates that longer flight durations are generally associated with higher prices, which makes intuitive sense as longer flights often cover greater distances or involve more complex routes.

*   **Price and Departure_Hour**: The correlation between `price` and `deparure_hour` (which represents the time of day) is **very weak (close to 0)**. This suggests that the numerical representation of departure hour itself does not have a strong linear relationship with the price. While specific departure *periods* (like 'Evening' or 'Early_Morning') might have different average prices, the hour as a continuous number doesn't show a strong linear correlation.

*   **Other Correlations**: Observe other correlations as well:
    *   `days_left` and `actual_duration` show a very weak positive correlation.
    *   `days_left` and `deparure_hour` show a very weak positive correlation.
    *   `actual_duration` and `deparure_hour` show a very weak negative correlation.

In summary, the most notable correlations affecting price are with `days_left` (negative) and `actual_duration` (positive), although both are relatively weak. `deparure_hour` as a numerical value does not show a significant linear correlation with price.
